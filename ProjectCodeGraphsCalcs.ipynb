```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
N = 64 #set lattice size
#%timeit
```

```python
#test individual c_L graphs
c_L = 3.5
y = np.loadtxt(str(c_L) + '.txt')
x = np.arange(1,len(y)+1)
plt.plot(x,y)
plt.title(r'Average density over time for $c_L =  %s$'%(c_L) + '\n' + r'and lattice size $%s^2$' %(N))
plt.xlabel('time [T]')
plt.ylabel(r'$E/N^2$')

#c_L=2.5, 150 iters, dt = 0.05 -> better convergence
#c_L=1.0, 150 iters, dt = 0.05 -> better convergence
#c_L=1.5, 150 iters, dt = 0.15 -> better convergence but not as smooth but E vs cL graph looks worse
```

```python
c_L = np.arange(0,5.5,0.5)
avgssE = np.zeros(len(c_L))
for i in range(0,len(c_L)):
    y = np.loadtxt(str(c_L[i]) + '.txt')
    x = np.arange(1,len(y)+1)
    avgssE[i] = y[-1]
    plt.figure(i)
    plt.plot(x,y)
    plt.title(r'Average density over time for $c_L =  %s$'%(c_L[i]) + '\n' + r'and lattice size $%s^2$' %(N))
    plt.xlabel('time [T]')
    plt.ylabel(r'$E/N^2$')

    
```

```python
plt.figure()
plt.plot(c_L,avgssE, "o")
plt.title(r'Average energy density against noise parameter for lattice size $%s^2$' %(N))
plt.xlabel(r'$c_L$')
plt.ylabel(r'$<E>/N^2$')

#first plot smoother - iterations 100 for all, dt = 0.05
```

```python
dEdcL = np.gradient(avgssE,0.5)

plt.figure()
plt.plot (c_L+0.25,dEdcL)
plt.title('Change in average energy density over change in noise \n' + r'against noise for lattice size $%s^2$' %(N))
plt.xlabel(r'$c_L$')
plt.ylabel(r'$d<E>/dc_L$')
```
